# handle-team-basics
This notebook combines two datasets (mens_basketball_programs and mens_institutions) into a single dataframe to be used in the future.

```python
# necessary imports
import pandas as pd
```

```python
# mens institutions
institutions = pd.read_csv('../data/csv/mens_institutions.csv')
institutions = institutions.sort_values('School')

# mens programs
programs = pd.read_csv('../data/csv/mens_basketball_programs.csv')
programs = programs.sort_values('School').reset_index(drop=True)
```

```python
programs["Tournament appearances"] = programs["Tournament appearances"].str.extract(r'(\d+)').astype(int)
programs["Final Four appearances"] = programs["Final Four appearances"].str.extract(r'(\d+)').astype(int)
programs["Championship wins"] = programs["Championship wins"].str.extract(r'(\d+)').astype(int)
```

```python
programs['School'] = programs['School'].replace(to_replace='\(.*\)', value='', regex=True).str.strip()
print(programs['School'].to_list())
```

```python
merged = pd.merge(programs, institutions, on='School', how='outer', suffixes=('_df1', '_df2'))
remove_columns = ["Nickname_df2", "Home arena"]
merged.drop(columns=remove_columns, inplace=True)
merged.rename(columns={'Nickname_df1': 'Nickname'}, inplace=True)
merged['ID'] = range(1, len(merged) + 1)
ordered_columns = ["ID", "School", "Common Name", "Nickname", "City", "State", "Type", "Subdivision", "Primary Conference", "Tournament appearances", "Final Four appearances", "Championship wins"]
merged = merged[ordered_columns]

merged
```

```python
merged.to_csv('../data/dataframes/team_info.csv', index=False)
```
